```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
import pandas as av
```

```python
url='https://www.flipkart.com/search?q=earbuds&sid=0pm%2Cfcn&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_3_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_3_na_na_na&as-pos=1&as-type=RECENT&suggestionId=earbuds%7CHeadphones&requestId=e6bbc6d1-e286-4d8b-848b-fa7e28aa4273&as-searchtext=ear'
```

```python
res=requests.get(url)
```

```python
res
```

```python
soup=BeautifulSoup(res.text)
```

```python
soup
```

```python
earbuds_names=soup.select('.s1Q9rs')
```

```python
earbuds_names
```

```python
earbuds_prices=soup.select('._8VNy32 ._30jeq3')
```

```python
earbuds_prices
```

```python
earbuds_offers=soup.select("._8VNy32 span")
```

```python
earbuds_offers
```

```python
df=av.DataFrame(earbuds_names,columns=['earbuds_names'])
```

```python
df
```

```python
df['earbuds_prices']=earbuds_prices
```

```python
df
```

```python
df['earbuds_offers']=earbuds_offers
```

```python
df
```

```python
df.to_csv('arivu.csv')
```
